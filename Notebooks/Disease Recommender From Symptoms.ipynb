```python
def ordinal_encoding(df,col):
    ordinal_encoder = ce.OrdinalEncoder(cols = [col],return_df = True)
    df_final = ordinal_encoder.fit_transform(df)
    return df_final
```

```python
df = ordinal_encoding(df,"prognosis")
```

```python
X = df.drop("prognosis",axis = 1)
Y = df["prognosis"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 27,test_size = 0.2)
```

```python
train,test = train_test_split(df,random_state = 27,test_size = 0.2)
```
